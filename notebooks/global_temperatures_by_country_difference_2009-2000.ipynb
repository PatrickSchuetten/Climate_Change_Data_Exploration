```python
import pandas as pd
import numpy as np

df=pd.read_csv('../data/cleaned/global_temperatures_by_country_2000_2009_cleaned.csv')
df
```

```python
df=df.drop('Unnamed: 0', axis=1)
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
filtered_df = df[~df['country'].isin(['Asia', 'Africa', 'North America', 'South America', 'Oceania', 'Europe'])]
filtered_df
```

```python
filtered_df['year'] = filtered_df['date'].dt.year
filtered_df
df_by_year_country = filtered_df.groupby(['country', 'year']).agg({'average_temperature': 'mean'}).reset_index()
df_by_year_country.head(20)
```

```python
df_by_year_country.to_csv("../data/cleaned/average_yearly_temperatures_by_country_2000_2009.csv")
```

```python
df_by_year_country_2000_2009 = df_by_year_country[df_by_year_country['year'].isin([2000,2009])]
df_by_year_country_2000_2009
```

```python
pivot_table = pd.pivot(df_by_year_country_2000_2009, index='country', columns='year', values='average_temperature').reset_index()
pivot_table
```

```python
pivot_table.columns = ['country', '2000', '2009']
pivot_table
```

```python
pivot_table['difference_in_temperature'] = pivot_table['2009'] - pivot_table['2000']
pivot_table
```

```python
pivot_table.to_csv("../data/cleaned/global_temperatures_by_country_difference_2009-2000_cleaned.csv")
```
