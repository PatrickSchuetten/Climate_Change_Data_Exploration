```python
import pandas as pd
import numpy as np

df=pd.read_csv('../data/cleaned/global_temperatures_by_country_1900_2012_cleaned.csv')
df
```

```python
df=df.drop('Unnamed: 0', axis=1)
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
filtered_df = df[~df['country'].isin(['Asia', 'Africa', 'North America', 'South America', 'Oceania', 'Europe'])]
filtered_df
```

```python
filtered_df['year'] = filtered_df['date'].dt.year
filtered_df
df_by_year_country = filtered_df.groupby(['country', 'year']).agg({'average_temperature': 'mean'}).reset_index()
df_by_year_country.head(20)
```

```python
df_by_year_country.to_csv("../data/cleaned/average_yearly_temperatures_by_country_2000_2009.csv")
```

```python
df_by_year_country_1900_2012 = df_by_year_country[df_by_year_country['year'].isin([1900,2012])]
df_by_year_country_1900_2012
```

```python
pivot_table = pd.pivot(df_by_year_country_1900_2012, index='country', columns='year', values='average_temperature').reset_index()
pivot_table
```

```python
pivot_table.columns = ['country', '1900', '2012']
pivot_table
```

```python
pivot_table['difference_in_temperature'] = pivot_table['2012'] - pivot_table['1900']
pivot_table
```

```python
pivot_table.to_csv("../data/cleaned/global_temperatures_by_country_difference_2012-1900_cleaned.csv")
```

```python
df2=pd.read_csv('../data/cleaned/global_temperatures_by_country_1980_2012_cleaned.csv')
df2=df2.drop('Unnamed: 0', axis=1)
df2
```

```python
df2['date']=pd.to_datetime(df2['date'])
df2.dtypes
```

```python
df2['year'] = df2['date'].dt.year
```

```python
df2_by_year = df2.groupby(['country', 'year']).agg({'average_temperature': 'mean'}).reset_index()
df2_by_year.head(20)
```

```python
df2_by_year_1980_2012 = df2_by_year[df2_by_year['year'].isin([1980,2012])]
df2_by_year_1980_2012
```

```python
pivot_table2 = pd.pivot(df2_by_year_1980_2012, index='country', columns='year', values='average_temperature').reset_index()
pivot_table2.columns = ['country', '1980', '2012']
pivot_table2['difference_in_temperature'] = pivot_table2['2012'] - pivot_table2['1980']
pivot_table2
```

```python
pivot_table2.to_csv("../data/cleaned/global_temperatures_by_country_difference_2012-1980_cleaned.csv")
```
